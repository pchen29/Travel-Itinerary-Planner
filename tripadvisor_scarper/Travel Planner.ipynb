### Hotel - Review Score

```python
def review_score_calculating(df):
    df_review = df[['n_Excellent', 'n_VeryGood', 'n_Average', 'n_Poor', 'n_Terrible']]
    total_score = df_review['n_Excellent'] * 5.0 + df_review['n_VeryGood'] * 4.0 \
               + df_review['n_Average'] * 3.0 + df_review['n_Poor'] * 2.0 \
               + df_review['n_Terrible'] * 1.0
    review_count = df_review['n_Excellent'] + df_review['n_VeryGood'] + df_review['n_Average']\
                 + df_review['n_Poor'] + df_review['n_Terrible']
    max_review_count = max(review_count)
    
    return (total_score / (review_count * 5.0) *(review_count / max_review_count)).tolist()

hotel_review_score = review_score_calculating(df_hotel)
```

```python
hotel_review_score
```

### Hotel - Amenity Property 

```python
amenities = df_hotel['Amenities']
all_amenity = {}
for amen in amenities:
    a_split = amen.split(',')
    for idx, a in enumerate(a_split):
        if idx < (len(a_split) - 1):
            term = a[2:-1]
            all_amenity[term] = all_amenity.get(term, 0) + 1
        else:
            term = a[2:-2]
            all_amenity[term] = all_amenity.get(term, 0) + 1
```

#### Extract high frequency amenities

```python
hf_amenities_tuple = sorted(all_amenity.items(), key = lambda all_amenity:all_amenity[1], reverse = True)[:20]
hf_amenities_tuple
```

```python
hf_amenities = [a[0] for a in hf_amenities_tuple]
hf_amenities
```

```python
# amenity property -> id
amen2id = {}
for idx, amen in enumerate(hf_amenities):
    amen2id[amen] = idx
    
# id -> amenity property 
id2amen = {}
for idx, amen in enumerate(hf_amenities):
    id2amen[idx] = amen
    
def amen2vec(user_amen, method2id):
    ua2vec = np.zeros(20)
    for ua in user_amen:
        idx = method2id[ua]
        ua2vec[idx] = 1
    return ua2vec

def preference_score(user_prefer, items_vec, method):
    # vectorize prefering items user chosen 
    prefer_vec = method(user_prefer)
    
    # get similarity
    similarity = prefer_vec * items_vec   
        
    # num of amenities user wanted
    total_ua = len(user_amen)
    
    return np.sum(similarity) / total_ua 

def preprocessing(am, hf):
    am_new = []
    a_split = am.split(',')
    for idx, a in enumerate(a_split):
        if idx < (len(a_split) - 1):
            term = a[2:-1]
            if term in hf:
                am_new.append(term)
        else:
            term = a[2:-2]
            if term in hf:
                am_new.append(term)
    return am_new
```

```python
ams = []
for a in amenities:
    a_tmp = preprocessing(a, hf_amenities)
    ams.append(a_tmp)
ams
```

```python
ams2vec = [amen2vec(a, amen2id) for a in ams]
```

### Hotel - Price

```python
def price_preprocessing(price_orig):
    price_processed = re.sub('[^0-9]', '', price_orig)
    return price_processed
```

```python
hotel_price = df_hotel['Price']
hotel_price_processed = [int(price_preprocessing(p)) for p in hotel_price]
```

```python
hotel_price_processed
```

```python
avg_price = np.sum(hotel_price_processed) / len(hotel_price_processed)
```

```python
hotel_price_score = avg_price / (avg_price + hotel_price_processed)
hotel_price_score
```
